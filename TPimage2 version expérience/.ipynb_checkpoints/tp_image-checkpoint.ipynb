### Ouverture et affichage d’une image
15’

Télécharger les images lena.tif et donjon_provins.tif sur le site.

Afficher les dans Python grâce aux commandes :

- I = plt.imread(...); I = I.astype(float) pour lire l’image. Le résultat est une matrice $N \times N$, qui code les niveaux de gris.

- trouver les valeurs maximale et minimale de niveau de gris et en déduire la profondeur d’encodage en nombre de bits à l’aide des fonctions max et min.

- afficher l’image par la commande : plt.imshow(I, cmap='gray'). 


```python
import numpy as np
from math import *
import matplotlib.pyplot as plt
I = plt.imread("lena.tif")
I = I.astype(float)
C = [[0 for j in range (1,len(I[0]) - 1 )] for k in range (1, len(I) - 1)]
for k in range (1, len(I) - 1) :
    for j in range (1, len(I[0]) - 1 ) :
        dx = ( I[k,j+1] - I[k,j-1] ) / 2
        dy = ( I[k+1,j] - I[k-1,j] ) / 2 
        C[k - 1][j - 1] = sqrt(dx*dx + dy*dy)


plt.imshow(C,cmap='gray')
plt.colorbar()


```

### Filtrage passe-bas de l’image
20’

Construire un noyau de filtre 2D séparable de la forme $h(x,y) = w(x)w(y)$ en utilisant le produit tensoriel de Python et la fonction hanning pour des valeurs faibles de longueur $L$ (typ. de 3 à 51). 

Rque : np.outer(vecteur, vecteur) = matrice. 

On normalisera le noyau de manière à avoir $\hat{h}(0,0) = 1$.

Filtrer l’image à l’aide de scipy.signal.convolve2d et afficher le résultat : 
filtered_image = sig.convolve2d(I, np.rot90(h), mode='same')

Quel est l’effet de ce filtrage ? Quel rôle joue le paramètre L ?


```python
import scipy.signal as sig
wx = np.hanning(10)
wy = wx 
h = np.outer(wx,wy) * (255/4500)
filtered_image = sig.convolve2d(I, np.rot90(h), mode='same')
plt.imshow(filtered_image,cmap='gray')
plt.colorbar()
```

### Contours
10’

Afficher l’image sc = s-sf où sf est l’image filtrée passe-bas. Interpréter rapidement le résultat. 



```python
s = I
sf = filtered_image
sc = s - sf
plt.imshow(sc,cmap='gray')
plt.colorbar()
```

### Amélioration de la netteté
20’

Une technique utilisée fréquemment pour amélioration de l’impression de netteté est d’utiliser l’image de contours pour accentuer le contraste local de l’image. Ici, on effectuera la simple addition pondérée : sn = s + a sc. Une valeur typique est a = 0.3. On prendra garde à tronquer les valeurs négatives de sn à 0 et > 255 à 255. 

Ex : sn [sn < 0] = 0 ; # met à zéro les valeurs negatives de sn

Afficher sn en jouant sur la valeur de a et de L.


```python
a = 0.3
sn = s + a*sc
sn [sn < 0] = 0
sn [sn > 255] = 255
plt.imshow(sn,cmap='gray')
plt.colorbar()
```

Effectuer le traitement précédent avec L=51 et a=1.2 sur l’image du donjon de Provins. 


```python
import scipy.signal as sig
wx = np.hanning(51)
wy = wx 
h = np.outer(wx,wy) * (255/120000)
filtered_image = sig.convolve2d(I, np.rot90(h), mode='same')
#plt.imshow(filtered_image,cmap='gray')
#plt.colorbar()
s = I
sf = filtered_image
sc = s - sf
plt.imshow(sc,cmap='gray')
plt.colorbar()
```
